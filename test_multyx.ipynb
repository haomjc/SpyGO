```python
# simulate gravitational law and demonstrate kepler's laws
import numpy as np
import matplotlib.pyplot as plt
import screwCalculus as sc
import casadi as ca
%matplotlib qt

# define the gravitational law
def grav_law(r, m1, m2):
    G = 1
    F = G*m1*m2/(r**2)
    return F

# define the initial conditions
r = 3 # distance between earth and sun
m1 = 6 # mass of earth
m2 = 1000# mass of sun
v = 0 # initial velocity
t = 0 # initial time
dt = 0.001 # time step

# simulate the gravitational law

# define the state variables
p1 = ca.SX.sym('p1',3,1) # position of earth
p2 = ca.SX.sym('p2',3,1) # position of sun
v1 = ca.SX.sym('v1',3,1) # velocity of earth
v2 = ca.SX.sym('v2',3,1) # velocity of sun

# define the initial conditions
p1_0 = np.array([r, 0, 0])
p2_0 = np.array([0, 0, 0])

# define the initial velocities
v1_0 = np.array([0, 10, 0])
v2_0 = np.array([0, 0, 0])

# define the initial state
x0 = np.concatenate((p1_0, p2_0, v1_0, v2_0))
print(x0)
# define the state variables
x = ca.vertcat(p1, p2, v1, v2)

# define the state equations
dist = ca.norm_2(p1-p2)
xdot = ca.vertcat(v1, v2, m2/(r**2)*(p2 - p1)/dist, m1/(r**2)*(p1 - p2)/dist)

# define the integrator
sol = sc.RK4(x, None, xdot, x0, t, 100, dt, None)

# animate the simulation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

plt.ion()
plt.show()
for i in range(sol.shape[1]):
    ax.scatter(sol[0, i], sol[1, i], sol[2, i], c='r')
    ax.scatter(sol[3, i], sol[4, i], sol[5, i], c='b')
    plt.pause(0.01)
    plt.draw()

# plot the simulation
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Material properties
E1, nu1 = 200e9, 0.3  # Material 1
E2, nu2 = 150e9, 0.28  # Material 2
F = 1000  # Normal force (N)

# Compute effective modulus
E_star = 1 / ((1 - nu1**2) / E1 + (1 - nu2**2) / E2)

# Define curvature radii
R1x, R1y = 0.02, 0.01  # Curvatures for body 1
R2x, R2y = 0.03, 0.015  # Curvatures for body 2

# Compute equivalent radii
Rx = 1 / (1/R1x + 1/R2x)
Ry = 1 / (1/R1y + 1/R2y)

# Compute semi-axes of contact ellipse
a = np.sqrt(3 * F * Rx / (4 * E_star))
b = np.sqrt(3 * F * Ry / (4 * E_star))

# Define grid points for visualization
N = 50
x_vals = np.linspace(-a, a, N)
y_vals = np.linspace(-b, b, N)
X, Y = np.meshgrid(x_vals, y_vals)

# Compute Hertzian pressure distribution with curvature
p0 = (3 * F) / (2 * np.pi * a * b)
P = p0 * np.sqrt(1 - (X**2 / a**2) - (Y**2 / b**2))
P[P < 0] = 0  # Ensure physically meaningful values

# Plot pressure distribution
plt.figure(figsize=(6, 5))
plt.contourf(X, Y, P, levels=30, cmap="coolwarm")
plt.colorbar(label="Pressure (Pa)")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("Hertzian Contact Pressure Distribution (Curved Surfaces)")
plt.show()
```
