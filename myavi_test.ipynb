```python
import pyvista as pv
import numpy as np
import time
# Set the notebook backend to 'client'
pv.set_jupyter_backend('trame')

# Create a plotter object
plotter = pv.Plotter()

# Create a sphere and add it to the plotter
sphere = pv.Sphere()
actor = plotter.add_mesh(sphere, color='cyan')

# Show the plotter in interactive mode (doesn't close after the first show)
plotter.show(auto_close=False)

# Define number of frames for the animation
n_frames = 100

# Loop over frames to rotate the sphere in real-time
for i in range(n_frames):
    # Rotate the sphere around the Z-axis
    actor.rotate_z(360 / n_frames)

    # Update the plotter to show the new frame
    plotter.update()

    # Allow PyVista to process events in the background
    plotter.app.processEvents()

    # Delay to control frame rate
    plotter.sleep(0.05)

# Close the plotter after the animation completes
plotter.close()

```
