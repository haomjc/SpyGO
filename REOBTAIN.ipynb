```python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays
COF_begin = [0.22, 0.25, 0.22, 0.27, 0.13, 0.18, 0.17, 0.19, 0.17, 0.18]
COF_end = [0.16, 0.17, 0.18, 0.14, 0.13, 0.14, 0.13, 0.16, 0.15, 0.16]
COF_avg = [0.17, 0.19, 0.17, 0.21, 0.13, 0.14, 0.14, 0.16, 0.15, 0.17]
COF_avg_std = [0.011, 0.016, 0.013, 0.046, 0.002, 0.004, 0.003, 0.006, 0.003, 0.008]
worn_area_normalized = np.array([0.96, 1.06, 1.34, 0.48, 1.11, 0.87, 0.97, 0.44, 1.47, 1.83])
worn_area_normalized = worn_area_normalized / worn_area_normalized.max()  # Normalize data

oil_names = ['Poliolo 70', 'Poliolo 100', 'Epossidato 70', 'Epossidato 100', 
             'HFAMES', 'Filtrato disidratato', 'Idrogenato', 'Estolide', 'C 18:1', 'Ferrari']

# Bar chart parameters
x = np.arange(len(oil_names))  # x-axis positions
width = 0.2  # Width of the bars

# Create the figure and two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot data on the left y-axis
bars1 = ax1.bar(x - width, COF_begin, width, label='COF Begin')
bars2 = ax1.bar(x, COF_end, width, label='COF End')
bars3 = ax1.bar(x + width, COF_avg, width, label='COF Average', color='orange', edgecolor='black')
ax1.errorbar(x + width, COF_avg, yerr=COF_avg_std, fmt='o', color='black', label='COF Avg Error')

# Customize the left y-axis
ax1.set_ylabel('Coefficient of Friction (COF)', color='blue')
ax1.set_ylim(0, 0.3)  # Adjust limits if needed
ax1.tick_params(axis='y', labelcolor='blue')
# ax1.legend(loc='upper left')

# Plot data on the right y-axis
ax2 = ax1.twinx()  # Create a second y-axis sharing the same x-axis
bars4 = ax2.bar(x + 2*width, worn_area_normalized, width, label='Normalized Worn Area', color='teal', edgecolor='black')

# Customize the right y-axis
ax2.set_ylabel('Normalized Worn Area', color='teal')
ax2.set_ylim(0, 1.2)  # Adjust limits if needed
ax2.tick_params(axis='y', labelcolor='teal')

# General plot settings
ax1.set_xticks(x)
ax1.set_xticklabels(oil_names, rotation=45, ha='right')
ax1.set_xlabel('Oil Names')
plt.title('COF and Normalized Worn Area with Dual Y-Axes')

# Add a legend for the second y-axis
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 0.95), ncol=4)
plt.tight_layout()
plt.grid(axis='y')

# Show the plot
plt.show()
x
```
